1. Simple ATM Calculator

```python
Balance = int(input("Enter your balance: "))
while True:
  Withdrawal = int(input("Enter your withdrawal amount: "))
  if Withdrawal > Balance:
    print("Insufficient balance")
  elif Balance > Withdrawal:
    Balance -=Withdrawal
    print(f"Transaction successful. Remaining balance: {Balance}")
    break
  else:
    print("Something went wrong!")
```

3. Temperature Converter

```python
Temp = int(input("Enter the temperature: "))
unit = input("Enter the unit (C/F): ").capitalize()
if unit == "C":
  Temp = (Temp*9/5)+32
  print(f"The temperature in Fahrenheit is: {Temp}")
elif unit == "F":
  Temp = (Temp-32)*5/9
  print(f"The temperature in Celsius is: {Temp}")
else:
  print("Something went wrong!")
```

6. Sum of even numbers

```python
number = int(input("Enter a number: "))
total = 0
for i in range(1, number + 1):
    if i % 2 == 0:
        total += i
print("Sum of even numbers:", total)
```

8. Dictionary – Word Frequency

```python
def Frequency():
  sentence = input("Enter a senternce")
  words = sentence.split()
  freq = {}
  for word in words:
    if word in freq:
      freq[word] += 1
    else:
      freq[word] = 1
  return freq
result = Frequency()
print(result)
```

8. Functions – Check Prime


```python
def check_prime(number):
  if number>1:
    for i in range(2,number):
      if number%i==0:
        print(number," is not a prime number")
        break
    else:
      print(number,"is prime number")
number=int(input("Enter a number"))
check_prime(number)
```

9. Vowel Counter

```python
def vowel():
  vowels = 'aeiouAEIOU'
  words = input("Enter a word to find the: ")
  count=0
  for word in words:
    if word in vowels:
      count += 1
  return count
result = vowel()
print(result)
```

10. Count Divisible Numbers

```python
ranges = int(input("Enter range for division: "))
divisor = int(input("Enter the divisor (D): "))
count = 0
for i in range(1, ranges + 1):
    if i % divisor == 0:
        count += 1
print(count)
```
