```python
import pandas as pd
def remove_outliers(data, column):
  data_filtered=data

  for i in column:
    Q1=data[i].quantile(0.25)
    Q3=data[i].quantile(0.75)
    IQR=Q3-Q1

    lower_whisker=Q1-1.5*IQR
    upper_whisker=Q3+1.5*IQR

    data_filtered=data_filtered[(data_filtered[i]>=lower_whisker) & (data_filtered[i]<=upper_whisker)]

  return data_filtered
```

```python
data=remove_outliers(data,column)
```

```python
from sklearn.preprocessing import LabelEncoder

catcolumns = ['Gender', 'Married', 'Education', 'Self_Employed', 'Loan_Status']
le = LabelEncoder()

# Apply the encoder to each column
for col in catcolumns:
    data[col] = le.fit_transform(data[col])
```

```python
data = pd.get_dummies(data, columns=['Property_Area'])

onehotcol = ['Property_Area_Rural', 'Property_Area_Semiurban', 'Property_Area_Urban']

for col in onehotcol:
    data[col] = data[col].astype(int)
```

```python
data['Dependents'] = data['Dependents'].replace(to_replace='3+', value='3')
data['Dependents'] = data['Dependents'].astype(int)
```
